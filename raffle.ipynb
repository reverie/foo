# Incomplete Control: Raffle odds calculator

## How many Fidenzas do you have?

```python
my_fidenzas = int(input())
```

## How many Crypto Citizens (Venetians + NYers) do you have?

```python
my_citizens = int(input())
```

```python

from math import factorial

def comb(n,r):
  return factorial(n) // factorial(r) // factorial(n-r)
    
NUM_FIDENZAS = 999

NUM_TICKETS = 24

not_my_fidenzas = NUM_FIDENZAS - my_fidenzas
ways_to_allocate = comb(NUM_FIDENZAS, NUM_TICKETS)

def ways_to_win_n(n):
  if n >= my_fidenzas:
    return 0
  losing_part = comb(not_my_fidenzas, NUM_TICKETS - n)
  winning_part = comb(my_fidenzas, n)
  return losing_part*winning_part

def prob_win_n(n):
  return ways_to_win_n(n) / ways_to_allocate

NUM_CITIZENS = 1500
NUM_CITIZEN_TICKETS = 23

not_my_citizens = NUM_CITIZENS - my_citizens
ways_to_allocate_citizens = comb(NUM_CITIZENS, NUM_CITIZEN_TICKETS)

def ways_to_win_n_citizens(n):
  if n >= my_citizens:
    return 0
  losing_part = comb(not_my_citizens, NUM_CITIZEN_TICKETS - n)
  winning_part = comb(my_citizens, n)
  return losing_part*winning_part

def prob_win_n_citizen(n):
  return ways_to_win_n_citizens(n) / ways_to_allocate_citizens

def prob_win_n_total(n):
  if n == 0:
    return prob_win_n(0)*prob_win_n_citizen(0)
  return sum(prob_win_n(f)*prob_win_n_citizen(n-f) for f in range(n+1))

for i in range(min(5, my_fidenzas + my_citizens)):
  print("Probability of winning {:d} Golden Token(s): {:.2f}%".format(i, prob_win_n_total(i)*100))

print("")
print("Overall chance you win: {:.2f}%".format(100 - 100*prob_win_n_total(0)))
```
